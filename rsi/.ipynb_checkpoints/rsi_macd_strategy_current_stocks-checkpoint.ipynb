## Analyze historical data to identify optimal stocks for swing trading.

```python
# List of stocks in the portfolio
 

swing_trading_stocks = [
    'AAPL', 'MSFT', 'NVDA', 'AMD', 'TSLA', 'GOOGL', 'AMZN',  # Technology Sector
    'NKE', 'SBUX', 'HD', 'DIS',                             # Consumer Discretionary
    'JNJ', 'PFE', 'ABBV', 'MRNA',                           # Healthcare
    'XOM', 'CVX', 'SLB',                                    # Energy
    'JPM', 'GS', 'BLK',                                     # Financials
    'AVGO', 'QCOM', 'INTC',                                 # Semiconductors
    'PG', 'KO', 'PEP',                                      # Consumer Staples
    'SPY', 'QQQ', 'IWM', 'TQQQ'                             # High Volatility ETFs
]

stocks = sp500_tickers

# Function to calculate RSI
def calculate_rsi(df, window=14):
    delta = df['Adj Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Function to calculate MACD
def calculate_macd(df, short_window=12, long_window=26, signal_window=9):
    short_ema = df['Adj Close'].ewm(span=short_window, adjust=False).mean()
    long_ema = df['Adj Close'].ewm(span=long_window, adjust=False).mean()
    macd = short_ema - long_ema
    signal = macd.ewm(span=signal_window, adjust=False).mean()
    macd_histogram = macd - signal
    return macd, signal

# Function to fetch and calculate all technical indicators
def fetch_and_calculate_data(stock):
    df = yf.download(stock, start="2023-01-01", end="2024-10-01")
    df['RSI'] = calculate_rsi(df)
    df['MACD'], df['Signal'] = calculate_macd(df)
    df['Volume_MA'] = df['Volume'].rolling(window=20).mean()  # 20-day moving average for volume
    return df

# Function to perform backtesting
def backtest(stock, data, investment=1):
    position = 0  # Start with no stock
    portfolio_value = 0
    cash = investment
    shares_held = 0
    initial_investment = investment
    
    for i in range(len(data)):
        if data['RSI'].iloc[i] < 30 and data['MACD'].iloc[i] > data['Signal'].iloc[i]:
            if cash > 0:
                # Buy signal
                shares_held += cash / data['Adj Close'].iloc[i]  # Buy shares
                cash = 0  # All cash invested
            
        elif data['RSI'].iloc[i] > 70 and data['MACD'].iloc[i] < data['Signal'].iloc[i]:
            if shares_held > 0:
                # Sell signal
                cash += shares_held * data['Adj Close'].iloc[i]  # Sell all shares
                shares_held = 0
        
        portfolio_value = cash + shares_held * data['Adj Close'].iloc[i]
    
    # Calculate return on investment
    roi = (portfolio_value - initial_investment) / initial_investment * 100
    return portfolio_value, roi

# Backtest the strategy for all stocks in the portfolio
results = {}

for stock in stocks:
    print(f"Fetching data and backtesting for {stock}...")
    data = fetch_and_calculate_data(stock)
    final_value, roi = backtest(stock, data)
    results[stock] = {'Final Value': final_value, 'ROI (%)': roi}

# Display the results
results_df = pd.DataFrame(results).T
results_df.sort_values(by='ROI (%)', ascending=False, inplace=True)
print(results_df)

# Plot the results
results_df['Final Value'].plot(kind='bar', title="Portfolio Final Value per Stock")
plt.ylabel('Final Value ($)')
plt.show()

# Cumulative ROI of the portfolio
cumulative_roi = results_df['ROI (%)'].mean()
print(f"\nCumulative Portfolio ROI: {cumulative_roi:.2f}%")
```

```python
# Import necessary libraries
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate RSI
def calculate_rsi(df, window=14):
    delta = df['Adj Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Function to calculate MACD
def calculate_macd(df, short_window=12, long_window=26, signal_window=9):
    short_ema = df['Adj Close'].ewm(span=short_window, adjust=False).mean()
    long_ema = df['Adj Close'].ewm(span=long_window, adjust=False).mean()
    macd = short_ema - long_ema
    signal = macd.ewm(span=signal_window, adjust=False).mean()
    return macd, signal

# Fetch data for TSLA
stock = 'TSLA'
df = yf.download(stock, start="2023-01-01", end="2024-10-01")

# Calculate RSI and MACD
df['RSI'] = calculate_rsi(df)
df['MACD'], df['Signal'] = calculate_macd(df)

# Drop rows with NaN values resulting from indicator calculations
df = df.dropna()

# Initialize buy and sell signals lists
buy_signals = [np.nan] * len(df)
sell_signals = [np.nan] * len(df)

# Buy and Sell signals based on RSI and MACD
for i in range(1, len(df)):
    if df['RSI'].iloc[i] < 30 and df['MACD'].iloc[i] > df['Signal'].iloc[i]:
        buy_signals[i] = df['Adj Close'].iloc[i]
    elif df['RSI'].iloc[i] > 70 and df['MACD'].iloc[i] < df['Signal'].iloc[i]:
        sell_signals[i] = df['Adj Close'].iloc[i]

# Add buy and sell signals to the DataFrame
df['Buy'] = buy_signals
df['Sell'] = sell_signals

# Plot stock price, volume, RSI, MACD, and Buy/Sell points
plt.figure(figsize=(14, 10))

# Plot Adjusted Close price and Buy/Sell signals
plt.subplot(4, 1, 1)
plt.plot(df['Adj Close'], label='Adjusted Close Price', color='blue', alpha=0.7)
plt.scatter(df.index, df['Buy'], color='green', label='Buy Signal', marker='^', alpha=1)
plt.scatter(df.index, df['Sell'], color='red', label='Sell Signal', marker='v', alpha=1)
plt.title(f'{stock} Adjusted Close Price with Buy/Sell Signals')
plt.legend()

# Plot Volume
plt.subplot(4, 1, 2)
plt.bar(df.index, df['Volume'], label='Volume', color='gray', alpha=0.6)
plt.title(f'{stock} Volume')
plt.legend()

# Plot RSI
plt.subplot(4, 1, 3)
plt.plot(df['RSI'], label='RSI', color='purple', alpha=0.7)
plt.axhline(30, linestyle='--', alpha=0.5, color='green')
plt.axhline(70, linestyle='--', alpha=0.5, color='red')
plt.title(f'{stock} RSI')
plt.legend()

# Plot MACD and Signal line
plt.subplot(4, 1, 4)
plt.plot(df['MACD'], label='MACD', color='blue', alpha=0.7)
plt.plot(df['Signal'], label='Signal Line', color='orange', alpha=0.7)
plt.fill_between(df.index, df['MACD'] - df['Signal'], 0, alpha=0.3, label='MACD Histogram')
plt.title(f'{stock} MACD')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Fetch data for ETH-USD
stock = 'ETH-USD'
df = yf.download(stock, start="2023-01-01", end="2024-10-01")

# Calculate RSI and MACD
df['RSI'] = calculate_rsi(df)
df['MACD'], df['Signal'] = calculate_macd(df)

# Calculate 3-day rolling average of RSI
df['RSI_3D_Avg'] = df['RSI'].rolling(window=3).mean()

# Drop rows with NaN values resulting from indicator calculations
df = df.dropna()

# Initialize buy and sell signals lists
buy_signals_r = [np.nan] * len(df)
sell_signals_r = [np.nan] * len(df)
buy_signals_m = [np.nan] * len(df)
sell_signals_m = [np.nan] * len(df)
buy_signals = [np.nan] * len(df)
sell_signals = [np.nan] * len(df)

# Buy and Sell signals based on RSI and MACD
for i in range(1, len(df)):
    if df['RSI'].iloc[i] < 30:
        buy_signals_r[i] = df['Adj Close'].iloc[i]
    elif df['RSI'].iloc[i] > 70:
        sell_signals_r[i] = df['Adj Close'].iloc[i]

# Loop through the DataFrame to find MACD Golden and Dead Cross
for i in range(1, len(df)):
    # Golden Cross (MACD line crosses above the Signal line)
    if df['MACD'].iloc[i] > df['Signal'].iloc[i] and df['MACD'].iloc[i - 1] <= df['Signal'].iloc[i - 1]:
        buy_signals_m[i] = df['Adj Close'].iloc[i]
    
    # Dead Cross (MACD line crosses below the Signal line)
    elif df['MACD'].iloc[i] < df['Signal'].iloc[i] and df['MACD'].iloc[i - 1] >= df['Signal'].iloc[i - 1]:
        sell_signals_m[i] = df['Adj Close'].iloc[i]

        

        # Loop through the DataFrame to find joint buy and sell signals
for i in range(1, len(df)):
    # Buy Signal: RSI rolling average <= 30 and MACD Golden Cross (MACD crosses above Signal)
    if df['RSI_3D_Avg'].iloc[i] <= 30 and df['MACD'].iloc[i] > df['Signal'].iloc[i] and df['MACD'].iloc[i - 1] <= df['Signal'].iloc[i - 1]:
        buy_signals[i] = df['Adj Close'].iloc[i]
    
    # Sell Signal: RSI rolling average >= 70 and MACD Dead Cross (MACD crosses below Signal)
    elif df['RSI_3D_Avg'].iloc[i] >= 70 and df['MACD'].iloc[i] < df['Signal'].iloc[i] and df['MACD'].iloc[i - 1] >= df['Signal'].iloc[i - 1]:
        sell_signals[i] = df['Adj Close'].iloc[i]

# Add buy and sell signals to the DataFrame
df['Buy RSI'] = buy_signals_r
df['Sell RSI'] = sell_signals_r
df['Buy MACD'] = buy_signals_m
df['Sell MACD'] = sell_signals_m
df['Buy Signal'] = buy_signals
df['Sell Signal'] = sell_signals    


# Plot stock price, volume, RSI, MACD, and Buy/Sell points
plt.figure(figsize=(14, 16))

# Plot Adjusted Close price and Buy/Sell signals
plt.subplot(6, 1, 1)
plt.plot(df['Adj Close'], label='Adjusted Close Price', color='blue', alpha=0.7)
plt.scatter(df.index, df['Buy RSI'], color='green', label='RSI Buy Signal', marker='P', alpha=1)
plt.scatter(df.index, df['Sell RSI'], color='red', label='RSI Sell Signal', marker='d', alpha=1)
plt.title(f'{stock} Adjusted Close Price with RSI Buy/Sell Signals')
plt.legend()


# Plot RSI
plt.subplot(6, 1, 2)
plt.plot(df['RSI'], label='RSI', color='purple', alpha=0.7)
plt.axhline(30, linestyle='--', alpha=0.5, color='green')
plt.axhline(70, linestyle='--', alpha=0.5, color='red')
plt.title(f'{stock} RSI')
plt.legend()

# Plot Adjusted Close price and Buy/Sell signals
plt.subplot(6, 1, 3)
plt.plot(df['Adj Close'], label='Adjusted Close Price', color='blue', alpha=0.7)
plt.scatter(df.index, df['Buy MACD'], color='blue', label='MACD Buy Signal', marker='o', alpha=1)
plt.scatter(df.index, df['Sell MACD'], color='orange', label='MACD Sell Signal', marker='s', alpha=1)
plt.title(f'{stock} Adjusted Close Price with MACD Buy/Sell Signals')
plt.legend()

# Plot MACD and Signal line
plt.subplot(6, 1, 4)
plt.plot(df['MACD'], label='MACD', color='blue', alpha=0.7)
plt.plot(df['Signal'], label='Signal Line', color='orange', alpha=0.7)
plt.fill_between(df.index, df['MACD'] - df['Signal'], 0, alpha=0.3, label='MACD Histogram')
plt.title(f'{stock} MACD')
plt.legend()

# Plot the Buy/Sell signals for MACD and RSI combination
plt.subplot(6, 1, 5)
plt.plot(df['Adj Close'], label='Adjusted Close Price', color='blue', alpha=0.6)
plt.scatter(df.index, df['Buy Signal'], color='green', label='Buy Signal (MACD + RSI)', marker='^', alpha=1)
plt.scatter(df.index, df['Sell Signal'], color='red', label='Sell Signal (MACD + RSI)', marker='v', alpha=1)
plt.title('Joint MACD + RSI Buy and Sell Signals with Adjusted Close Price')
plt.legend()

# Plot Volume
plt.subplot(6, 1, 6)
plt.bar(df.index, df['Volume'], label='Volume', color='gray', alpha=0.6)
plt.title(f'{stock} Volume')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Import necessary libraries
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate RSI
def calculate_rsi(df, window=14):
    delta = df['Adj Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Function to calculate MACD
def calculate_macd(df, short_window=12, long_window=26, signal_window=9):
    short_ema = df['Adj Close'].ewm(span=short_window, adjust=False).mean()
    long_ema = df['Adj Close'].ewm(span=long_window, adjust=False).mean()
    macd = short_ema - long_ema
    signal = macd.ewm(span=signal_window, adjust=False).mean()
    return macd, signal

# Fetch data for TSLA
stock = 'TSLA'
df = yf.download(stock, start="2023-01-01", end="2024-10-17")

# Calculate RSI and MACD
df['RSI'] = calculate_rsi(df)
df['MACD'], df['Signal'] = calculate_macd(df)

# Calculate 3-day rolling average of RSI
df['RSI_3D_Avg'] = df['RSI'].rolling(window=3).mean()

# Drop rows with NaN values resulting from indicator calculations
df = df.dropna()

# Initialize buy and sell signals lists
buy_signals = [np.nan] * len(df)
sell_signals = [np.nan] * len(df)
    

# Loop through the DataFrame to find joint buy and sell signals
for i in range(1, len(df)):
    # Buy Signal: RSI rolling average <= 30 and MACD Golden Cross (MACD crosses above Signal)
    if df['RSI_3D_Avg'].iloc[i] <= 35 and df['MACD'].iloc[i] > df['Signal'].iloc[i] and df['MACD'].iloc[i - 1] <= df['Signal'].iloc[i - 1]:
        buy_signals[i] = df['Adj Close'].iloc[i]
    
    # Sell Signal: RSI rolling average >= 70 and MACD Dead Cross (MACD crosses below Signal)
    elif df['RSI_3D_Avg'].iloc[i] >= 70 and df['MACD'].iloc[i] < df['Signal'].iloc[i] and df['MACD'].iloc[i - 1] >= df['Signal'].iloc[i - 1]:
        sell_signals[i] = df['Adj Close'].iloc[i]

# Add buy and sell signals to the DataFrame
df['Buy Signal'] = buy_signals
df['Sell Signal'] = sell_signals    


# Plot stock price, volume, RSI, MACD, and Buy/Sell points
plt.figure(figsize=(14, 10))


# Plot the Buy/Sell signals for MACD and RSI combination
plt.subplot(4, 1, 1)
plt.plot(df['Adj Close'], label='Adjusted Close Price', color='blue', alpha=0.6)
plt.scatter(df.index, df['Buy Signal'], color='green', label='Buy Signal (MACD + RSI)', marker='^', alpha=1)
plt.scatter(df.index, df['Sell Signal'], color='red', label='Sell Signal (MACD + RSI)', marker='v', alpha=1)
plt.title(f'{stock} Joint MACD + RSI Buy and Sell Signals with Stock Price')
plt.legend()

# Plot MACD and Signal line
plt.subplot(4, 1, 2)
plt.plot(df['MACD'], label='MACD', color='blue', alpha=0.7)
plt.plot(df['Signal'], label='Signal Line', color='orange', alpha=0.7)
plt.fill_between(df.index, df['MACD'] - df['Signal'], 0, alpha=0.3, label='MACD Histogram')
plt.title(f'{stock} MACD')
plt.legend()

# Plot RSI
plt.subplot(4, 1, 3)
plt.plot(df['RSI'], label='RSI', color='purple', alpha=0.7)
plt.axhline(30, linestyle='--', alpha=0.5, color='green')
plt.axhline(70, linestyle='--', alpha=0.5, color='red')
plt.title(f'{stock} RSI')
plt.legend()


# Plot Volume
plt.subplot(4, 1, 4)
plt.bar(df.index, df['Volume'], label='Volume', color='gray', alpha=0.6)
plt.title(f'{stock} Volume')
plt.legend()

plt.tight_layout()
plt.show()
```
